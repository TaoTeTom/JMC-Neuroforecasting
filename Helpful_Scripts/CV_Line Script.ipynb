```python
from urllib.parse import urljoin
import pandas as pd
from urllib.request import urlopen
import os
```

```python
##Splits Dataframe by candidate
school_list = pd.read_csv(r"C:\Users\15the\OneDrive - California Institute of Technology\Selenium_JMC\Schools_List.csv")


candidaes_that_didnt_work = []
for school in school_list.values:
    school_folders = []
    path = fr"C:\Users\15the\OneDrive - California Institute of Technology\Selenium_JMC\{school[0]}"
    print(path)
    school_folders.append(path)
    for folder in school_folders:
        os.chdir(folder)
        excel = fr"{folder}\{school[0]}.csv"
        candidate_list_df = pd.read_csv(excel, engine='python')
        for index, row in candidate_list_df.iterrows():
            name = f"{row['First Name']} {row['Last Name']}"
            candidate_folder = fr"{folder}\{name}"
            try:
                os.chdir(candidate_folder)
            except:
                print(f"{candidate_folder} does not have any candidates")
                candidaes_that_didnt_work.append((school[0], name))
                continue

            row_df = row.to_frame().T

            row_df.to_csv(f"{name}.csv", index=False)
print(candidaes_that_didnt_work)

```
